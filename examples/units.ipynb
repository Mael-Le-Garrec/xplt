# Units and Labels

The library comes with default settings for units, labels and symbols as used by [Xsuite](https://xsuite.readthedocs.io). You can add or change the units, labels and symbols associated with properties with the following methods.

```python
import numpy as np
import xplt
```

```python
# custom data
data = {
    "x": np.random.normal(size=100),  # in units of m  --> use library default
    "y": np.random.normal(size=100),  # in units of mm --> overwrite library default below
    "a": np.random.normal(size=100),  # in units of s  --> register unit below
}
```

## Global settings for all plots

To register or change the units associated with the raw data of a property use:

```python
xplt.register_property("y", "mm", symbol="$\\upsilon$")
xplt.register_property("a", "s", description="Age")
```

```python
plot = xplt.PhaseSpacePlot(data, "x-y,a-y")
```

## Settings for a single plot

Use the `data_units` and `display_units` keyword arguments:

```python
plot = xplt.PhaseSpacePlot(
    data, "x-y,a-y", data_units={"y": "mm", "a": "ms"}, display_units={"y": "cm"}
)
```
